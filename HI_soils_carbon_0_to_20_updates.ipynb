```python
# List of soil_column_ids to remove
remove_ids = [
    'NCSS_29901', 'NCSS_20949', 'NCSS_21776', 'NCSS_5303',
    'NCSS_20946', 'NCSS_21530', 'NCSS_16134', 'NCSS_16137',
    'NCSS_16136', 'NCSS_28765', 'NCSS_21370', 'NCSS_25632', 'NCSS_25633'
]

# Remove them from the DataFrame
soc2025 = soc2025[~soc2025['soil_column_id'].isin(remove_ids)]
```

```python
# Step 1: Filter to only NCSS rows with 010 or 1030 cm data
subset = soc2025[
    (soc2025['source_dataset'] == 'NCSS') &
    (((soc2025['depth_top'] == 0) & (soc2025['depth_adj_bottom'] == 10)) |
     ((soc2025['depth_top'] == 10) & (soc2025['depth_adj_bottom'] == 30)))
].copy()

# Step 2: Create unique location ID (soil_column_id is best here)
subset['loc_id'] = subset['soil_column_id']

# Step 3: Group and impute
def impute_0_20(group):
    if group.shape[0] < 2:
        return None
    d = dict(zip(group['depth_top'], group['c_perc']))
    if 0 in d and 10 in d:
        return (d[0] * 10 + d[10] * 10) / 20
    return None

# Apply the imputation
imputed_vals = subset.groupby('loc_id').apply(impute_0_20).dropna()
imputed_vals.name = 'c_perc_0_20'

# Step 4: Assign back to soc2025, only for matching NCSS soil_column_ids
soc2025['c_perc_0_20'] = soc2025.apply(
    lambda row: imputed_vals[row['soil_column_id']] if row['source_dataset'] == 'NCSS' and row['soil_column_id'] in imputed_vals else None,
    axis=1
)
```

```python
# Step 1: Filter to NCSS rows that have 010 cm depth
ncss_0_10 = soc2025[
    (soc2025['source_dataset'] == 'NCSS') &
    (soc2025['depth_top'] == 0)
].copy()

# Step 2: Keep only the columns you'll reuse
columns_to_keep = soc2025.columns  # keeps everything as is

# Step 3: Create the new 020 cm row
new_rows = ncss_0_10.copy()

# Update depths
new_rows['depth_top'] = 0
new_rows['depth_bottom'] = 20
new_rows['depth_adj_bottom'] = 20

# Assign imputed values from c_perc_0_20
new_rows['c_perc'] = new_rows['c_perc_0_20']
new_rows['imp_c'] = new_rows['c_perc_0_20']

# Optional: remove unique_id or other fields you dont want duplicated
# new_rows['unique_id'] = np.nan

# Step 4: Append the new rows to soc2025
soc2025 = pd.concat([soc2025, new_rows], ignore_index=True)
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
```

```python
# Step 1: Load the shapefile
island_gdf = gpd.read_file('zip:///content/HI_island_boundaries.zip')
```

```python
print(island_gdf.columns)
```

```python
# Step 2: Convert soc2025 to a GeoDataFrame
# Filter rows where island is missing and lat/lon are present
missing_island = soc2025[soc2025['island'].isna() & soc2025['latitude'].notna() & soc2025['longitude'].notna()].copy()

# Create Point geometries
missing_island['geometry'] = missing_island.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
missing_island_gdf = gpd.GeoDataFrame(missing_island, geometry='geometry', crs='EPSG:4326')
```

```python
# Step 3: Ensure both layers are in the same CRS
if island_gdf.crs != 'EPSG:4326':
    island_gdf = island_gdf.to_crs('EPSG:4326')

# Step 4: Spatial join to get island name
joined = gpd.sjoin(missing_island_gdf, island_gdf[['geometry', 'Island']], how='left', predicate='within')

# Step 5: Fill in the island name back into soc2025
soc2025.loc[joined.index, 'island'] = joined['Island'].values

print(soc2025['island'].isna().sum())
```
